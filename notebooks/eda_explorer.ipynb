Now, we will create our Data Dictionary, as it orders the instructiones for this first advance of the heavy Big Data project.

```python
# We convert values in column marcador to numeric instead of object type
if "marcador" in df_copy.columns:
        goles_split = df_copy["marcador"].str.extract(r'(\d+)\D+(\d+)')
        df_copy["home_goals"] = pd.to_numeric(goles_split[0], errors="coerce")
        df_copy["away_goals"] = pd.to_numeric(goles_split[1], errors="coerce")
```

```python
# A better way to manage some null values that did not die after the deleteing of some
df_copy = df_copy.fillna({
        "entrenador_local": "unknown",
        "entrenador_visitante": "unknown",
        "goles": "unknown",
        "global": "not_applicable"
    })
```

```python
def create_data_dictionary(df: pd.DataFrame) -> pd.DataFrame:
    descriptions = {
        "season": "Season of the competition",
        "fase": "Competition phase (group stage, knockout)",
        "instancia": "Specific round (e.g., quarter-final)",
        "fecha": "Match date",
        "local": "Home team",
        "visitante": "Away team",
        "marcador": "Match score (home-away)",
        "home_goals": "Goals scored by home team",
        "away_goals": "Goals scored by away team",
        "global": "Aggregate score in knockout rounds",
        "estadio": "Stadium name",
        "ciudad": "City of the stadium",
        "pais": "Country where match was played",
        "entrenador_local": "Home team coach",
        "entrenador_visitante": "Away team coach",
        "goles": "Goal details or scorers"
    }

    data_dict = pd.DataFrame({
        "column": df.columns,
        "dtype": df.dtypes.astype(str),
        "missing": df.isnull().sum(),
        "missing_%": (df.isnull().mean() * 100).round(2),
        "description": [descriptions.get(col, "No description") for col in df.columns],
        "example": [str(df[col].dropna().iloc[0]) if df[col].dropna().shape[0] > 0 else "" for col in df.columns]
    })

    return data_dict
```

```python
print("📚 Generating data dictionary...")
data_dict = create_data_dictionary(df_copy)
data_dict.to_csv("../data/processed/data_dictionary_v1.csv", index=False)
```
